**Data Preprosseing**

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('laptop_price.csv',encoding='latin-1')
```

First we need to check are there any column we can simply remove(no effect for the price) > here we have id coloumn 

```python
data = data.drop('laptop_ID',axis='columns')
```

```python
data.head()
```

Now we need to check the object columns and convert those into numerical types

Here we have ram and weight column , we can simply remove kg and GB parts from those and convert to numerical types

```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')
data.info()
```

Now we should check other object columns one by one , and identify different types of each column have , if there are more types we have to reduce the types 

```python
data['Company'].value_counts()
```

here we have so many types if we do the one hot encoding there are 19 columns , so we get the all types together having value counts lower than 10

```python
def company_categorize(company):
    if data['Company'].value_counts()[company] < 10:
        return 'Other'
    else:
        return company

data['Company'] = data['Company'].apply(company_categorize)
```

Now we check product column

```python
data['Product'].value_counts()
len(data['Product'].value_counts())
```

here there are 618 types so so product should differ almost over every column so , so it does not effect price more so we can simply remove the column

```python
data=data.drop('Product',axis='columns')
```

now we check TypeName column 

```python
data['TypeName'].value_counts()
```

here it has only six so for one hot encoding it does not matter so we can keep it without changing

Inches is a numerical column , so we move to resolution column

```python
data['ScreenResolution'].value_counts()
```

here there are lot of types , so I check weather it has ips pannel and touch screen or not as a new column and remove this Screen resolution column 

```python
data['TouchScreen']=data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)

data['IPS'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data=data.drop('ScreenResolution',axis='columns')
```

Cpu cloumn

```python
data['Cpu'].value_counts()
```

now we get first 3 words to initalize the cpu type

```python
data['Cpu_type']=data['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
data['Cpu_type'].value_counts()
```

now we get types intel core i3 i5 i7 and amd and other to reduce types

```python
def cpu_categorize(cpu):
    if cpu == 'Intel Core i7' or cpu == 'Intel Core i5' or cpu == 'Intel Core i3':
        return cpu
    else:
        if cpu.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'

data['Cpu_type']=data['Cpu_type'].apply(cpu_categorize)
```

```python
data['Cpu_type'].value_counts()
```

```python
data=data.drop('Cpu',axis='columns') 
```

Now we move to the gpu column 

```python
data['Gpu'].value_counts()
```

We only categorize Intel , AMD , And Nvidea

```python
data['Gpu_brand']=data['Gpu'].apply(lambda x:x.split()[0])
```

```python
data['Gpu_brand'].value_counts()
```

```python
#here we have other type ARM we simply remove it
data=data[data['Gpu_brand']!='ARM']
```

```python
data=data.drop('Gpu',axis='columns')
```

```python
data['OpSys'].value_counts()
```

```python
def os_categorize(os):
    if os=='Windows 10' or os=='Windows 7' or os=='Windows 10 S':
        return 'Windows'
    elif os=='Mac OS X' or os=='macOS':
        return 'Mac'
    elif os=='Linux' :
        return os
    else:
        return 'Other'
data['Opsys_type']=data['OpSys'].apply(os_categorize)
```

```python
data['Opsys_type'].value_counts()
```

```python
data=data.drop('OpSys',axis='columns')
```

```python
data = data.drop('Inches',axis='columns')
```

```python
data.head()
```

Now we initialize all columns that can be use to one hot encoding 

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.shape
```

**Model Building**

```python
x = data.drop('Price_euros',axis='columns')
y = data['Price_euros']
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def model_acc(model):
    acc = cross_val_score(model,x,y,cv=5).mean()
    print(str(model)+' --> '+str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_acc(lr)


from sklearn.linear_model import Lasso
lasso=Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_acc(rf)
```

here we can identify that RandomForest algorithm has the highest accuracy

**Hyper parameter tuning**

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators' : [10,50,100],
              'criterion' : ['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf,param_grid=parameters)

grid_fit = grid_obj.fit(x_train,y_train)
best_rf = grid_fit.best_estimator_
best_rf
```

```python
best_rf.score(x_test,y_test)
```

**Saving the model for future use using pickle**

```python
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_rf,file)
```
